```python
# re-establish datetime vars..csv format doesn't save datetimes (it does preserve numeric formats anc object)
df['Trade Date'] = pd.to_datetime(df['Trade Date'])
df['opFOM'] = pd.to_datetime(df['opFOM'])
df['t01XP'] = pd.to_datetime(df['t01XP'])
df['LTD'] = pd.to_datetime(df['LTD'])
df['dt01'] = pd.to_datetime(df['dt01'])
df['dt02'] = pd.to_datetime(df['dt02'])
df['opXP'] = pd.to_datetime(df['opXP'])
```

```python
#df['numMthM01'] = np.where(df['LTD']- df['Trade Date']> 0, df['opMth'], df['calMth'])
## this doesn't work because date difference is being compared to an int 
## instead create datediff column
```

```python
df['diffDates'] = df['LTD'] - df['Trade Date']  
## shows us a zero if the contract has just started/ended
## a negative number means the calendar month does not equal the current futures month
## the negative can be thought of as how many days into the (whatever the next month is from the calendar month) contract
## for example:
## if the trade date is 6/22/2020, the spx next month (7) options expires on Friday 7/17
## therefore the VIX futures (expiring 30 days before the spx Options) expired already on 6/17
## the diffDates says -5 because we are not in the 6th month futures, now the front month contract is July(7th month)
## so on 6/22/2020, we are 5 days into the new front month (term 01) contract of July
```

```python
## futures months straddle two calendar months, so the futures month dosen't always equal the calendar month 
## (see above)
df['numMthT01'] = np.where(df['diffDates'].dt.days >= 0, df['calMth'], df['opMth'])
```
