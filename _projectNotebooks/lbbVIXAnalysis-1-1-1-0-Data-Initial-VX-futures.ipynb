## STEP ONE -- JUST GET THE DATA

```python
##IMPORT OF RAW DIRTY DATA FROM CBOE FROM THE FAR PAST
##2004 - 2018


def main(url):
    with requests.Session() as req:
        r = req.get(url)
        soup = BeautifulSoup(r.content, 'html.parser')
        target = [f"{url[:20]}{item['href']}" for item in soup.select(
            "a[href$='VX.csv']")]
        for x in target:
            print(f"Downloading {x}")
            r = req.get(x)
            name = x.rsplit("/", 1)[-1]
            with open(name, 'wb') as f:
                f.write(r.content)


main("https://www.cboe.com/products/futures/market-data/historical-data-archive")

## from this link https://stackoverflow.com/questions/62019059/downloading-multiple-csv-files-from-a-webpage-using-
## beautifulsoup

```
