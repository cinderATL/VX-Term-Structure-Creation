## Finally! Merge all cleaned csv files into one file

```python
##--- constant errors made it impossible to get these files merged together, I finally figured
## out how to tell me there were bad lines in the first place, and upon finding them, don't 
## hold up the merge operations, just skip. Bad lines are usually those lines with too many fields.

path = r'C:\Users\cinderella\dataSci\DataScience\eliteDataScience\07-CapstoneProject\Final\Data-LATEST-VX' # use your path
all_files = glob.glob(os.path.join(path, "*.csv"))  
li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0, keep_default_na=True, error_bad_lines = False, warn_bad_lines = False)
    li.append(df)

bigDFframe = pd.concat(li, axis=0, ignore_index=True)
```
